```python
import datetime
date_time = datetime.datetime.now()
car['Age']=date_time.year - car['Year']
car.head()
```

```python
car.drop('Year',axis=1,inplace=True)
```

```python
car.head()
```

```python
sns.boxplot(car['Selling_Price'])
```

```python
sorted(car['Selling_Price'],reverse=True)
```

```python
car = car[~(car['Selling_Price']>=33.0) & (car['Selling_Price']<=35.0)]
```

```python
car.shape
```

```python
car.head(1)
```

```python
car['Fuel_Type'].unique()
```

```python
car['Fuel_Type'] = car['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
```

```python
car['Fuel_Type'].unique()
```

```python
car['Selling_type'].unique()
```

```python
car['Selling_type'] = car['Selling_type'].map({'Dealer':0,'Individual':1})
```

```python
car['Selling_type'].unique()
```

```python
car['Transmission'].unique()
```

```python
car['Transmission'] =car['Transmission'].map({'Manual':0,'Automatic':1})
car['Transmission'].unique()
```

```python
car.head()
```

```python
X = car.drop(['Car_Name','Selling_Price'],axis=1)
y = car['Selling_Price']
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)

rf = RandomForestRegressor()
rf.fit(X_train,y_train)

xgb = GradientBoostingRegressor()
xgb.fit(X_train,y_train)

xg = XGBRegressor()
xg.fit(X_train,y_train)
```

```python
y_pred1 = lr.predict(X_test)
y_pred2 = rf.predict(X_test)
y_pred3 = xgb.predict(X_test)
y_pred4 = xg.predict(X_test)
```

```python
from sklearn import metrics
score1 = metrics.r2_score(y_test,y_pred1)
score2 = metrics.r2_score(y_test,y_pred2)
score3 = metrics.r2_score(y_test,y_pred3)
score4 = metrics.r2_score(y_test,y_pred4)
```

```python
print(score1,score2,score3,score4)
```

```python
final_data = pd.DataFrame({'Models':['LR','RF','GBR','XG'],
             "R2_SCORE":[score1,score2,score3,score4]})
```

```python
final_data
```

```python
sns.barplot(x=final_data['Models'], y=final_data['R2_SCORE'])
```

```python
import joblib
import pandas as pd

def predict_car_price(p1, p2, fuel_type, Seller_Type, Transmission, p6, p7):
    model = joblib.load('car_price_predictor')

    # Create a DataFrame with one-hot encoding for categorical variables
    data_new = pd.DataFrame({
        'Present_Price': [p1],
        'Kms_Driven': [p2],
        'Fuel_Type': [fuel_type],  # Assuming 'Fuel_Type' was a feature during training
        'Selling_type': [Seller_Type],  # Assuming 'Selling_type' was a feature during training
        'Transmission': [Transmission],  # Assuming 'Transmission' was a feature during training
        'Owner': [p6],
        'Age': [p7]
    })

    result = model.predict(data_new)
    return result[0]

print("Car Price Prediction Using Machine Learning")

p1 = float(input("Enter Present_Price: "))
p2 = float(input("Enter Kms_Driven: "))
fuel_type = input("Enter Fuel_Type (e.g., petrol, diesel, cng): ")
Seller_Type = input("Enter Seller_Type (e.g., dealer, individual): ")
Transmission = input("Enter Transmission (e.g., manual, automatic): ")
p6 = float(input("Enter Owner: "))
p7 = float(input("Enter Age of the car in years:"))

prediction = predict_car_price(p1, p2, fuel_type, Seller_Type, Transmission, p6, p7)
print("Car Purchase amount:", prediction)
```
