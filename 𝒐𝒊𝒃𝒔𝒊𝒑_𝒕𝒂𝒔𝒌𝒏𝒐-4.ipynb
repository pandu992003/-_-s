```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python

encodings_to_try = ['utf-8', 'latin1', 'ISO-8859-1', 'cp1252']

for encoding in encodings_to_try:
    try:
        spam = pd.read_csv('/content/spam.csv', encoding=encoding)
        break  # Break the loop if successful
    except UnicodeDecodeError:
        continue  # Continue to the next encoding if decoding fails

```

```python
print(spam)
```

```python
# replace the null values with a null string
mail_data = spam.where((pd.notnull(spam)),'')
```

```python
# printing the first 5 rows of the dataframe
mail_data.head()
```

```python
# checking the number of rows and columns in the dataframe
mail_data.shape
```

```python
last_two_columns = spam.iloc[:, :2]
```

```python
# checking the number of rows and columns in the dataframe
last_two_columns.shape
```

```python
# printing the first 5 rows of the dataframe
last_two_columns.head()
```

```python
last_two_columns.info()
```

```python
# label spam mail as 0;  ham mail as 1;

last_two_columns.loc[last_two_columns['v1'] == 'spam', 'v1',] = 0
last_two_columns.loc[last_two_columns['v1'] == 'ham', 'v1',] = 1
```

```python
# separating the data as texts and label

X = last_two_columns['v2']

Y = last_two_columns['v1']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
# training the Logistic Regression model with the training data
model.fit(X_train_features, Y_train)
```

```python
# prediction on training data

prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
# prediction on test data

prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

```python
input_mail = ["I've been searching for the right words to thank you for this breather. I promise i wont take your help for granted and will fulfil my promise. You have been wonderful and a blessing at all times"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

# making prediction

prediction = model.predict(input_data_features)
print(prediction)


if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
```
